```python
import numpy as np

def jacobi(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
import numpy as np

def gauss_seidel(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
import numpy as np

# 6a

A_a = np.array([
    [10,-1, 2, 0, 0],
    [-1,11,-1, 3, 0],
    [2, -1,10,-1, 0],
    [0,  3,-1, 8,-2],
    [0,  0, 0,-2, 9]
])

b_a = np.array([14,30,26,25,37])
```

```python
# 6b

A_b = np.array([
    [1, 2, 3, 0, 0],
    [2, 1, 2, 3, 0],
    [3, 2, 1, 2, 3],
    [0, 3, 2, 1, 2],
    [0, 0, 3, 2, 1]
])

b_b = np.array([14, 22, 33, 26, 22])
```

```python
# jacobi method

import time

# for 6a

x0 = np.zeros_like(b_a)

start = time.time()

solution, iterations = jacobi(A_a, b_a, x0)
end = time.time()

if (iterations != 100):    
    print(f"[6a] Jacobi Solution: {solution}")
    print(f"[6a] Jacobi Iterations: {iterations}")
    print(f"[6a] Execution took: {end - start} s\n")
else:
    print("[6a] No solution found within 100 iterations")


# for 6b

x0 = np.zeros_like(b_b)

start = time.time()

solution, iterations = jacobi(A_b, b_b, x0)
end = time.time()

if (iterations != 100):    
    print(f"[6b] Jacobi Solution: {solution}")
    print(f"[6b] Jacobi Iterations: {iterations}")
    print(f"[6b] Execution took: {end - start} s")
else:
    print("[6b] No solution found within 100 iterations")
```

```python
# Gauss Seidel method

import time

# for 6a

x0 = np.zeros_like(b_a)

start = time.time()

solution, iterations = gauss_seidel(A_a, b_a, x0)
end = time.time()

if (iterations != 100):    
    print(f"[6a] GS Solution: {solution}")
    print(f"[6a] GS Iterations: {iterations}")
    print(f"[6a] Execution took: {end - start} s\n")
else:
    print("[6a] No solution found within 100 iterations")


# for 6b

x0 = np.zeros_like(b_b)

start = time.time()

solution, iterations = gauss_seidel(A_b, b_b, x0)
end = time.time()

if (iterations != 100):    
    print(f"[6b] GS Solution: {solution}")
    print(f"[6b] GS Iterations: {iterations}")
    print(f"[6b] Execution took: {end - start} s")
else:
    print("[6b] No solution found within 100 iterations")
```
