```python
import random
```

```python
# monte carlo

def a_pi(n):
    inside = 0
    for _ in range(0, n):
        inside += int((random.uniform(-1,1)**2 + random.uniform(-1,1)**2)**(1/2) < 1)
            
    return (4*inside/n)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#driver code

# create axes
x = np.arange(0, 100)

y_abs = np.empty(100)
y_rel = np.empty(100)

for i in range(0, 100):
    # fill error values
    calc = a_pi(1000)
    y_abs[i] = abs(calc - np.pi)
    y_rel[i] = abs((calc - np.pi) / np.pi)    
```

```python
# plotting

plt.subplot(2,2,1) # (row, col, active_area)
plt.title("absolute error")
plt.plot(x, y_abs, 'r')

plt.subplot(2,2,2)
plt.title("relative error")
plt.plot(x, y_rel, 'g')
```
