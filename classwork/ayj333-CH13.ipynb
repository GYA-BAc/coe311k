```python
import numpy as np
```

```python
def mat_mul(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
                
    return C
```

```python
import time


def timed_run(f):
    start = time.time()
    f()
    end = time.time()
    return end-start


ns = np.empty(8)
naive = np.empty(8)
optim = np.empty(8)

n = 3
for i in range(8):
    A = np.random.randint(0, 10, (n, n))
    B = np.random.randint(0, 10, (n, n))
    
    ns[i] = n
    
    naive[i] = timed_run(lambda: mat_mul(A, B))
    optim[i] = timed_run(lambda: A@B)
    
    n *= 2
    
print("Done.")
```

```python
import matplotlib.pyplot as plt

plt.plot(ns, naive, label="naive")
plt.plot(ns, optim, label="optimized")
plt.legend()
plt.show()
```
