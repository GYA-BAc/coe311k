```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X, Y)
plt.plot(np.linspace(0, 5, 5), np.linspace(0, 5, 5)*m + c)
plt.show()
```

```python
# with randomized vectors
X = np.random.uniform(0, 5, 10)
Y = np.random.uniform(0, 5, 10)

# Stack the design matrix with a column of ones (for the intercept)

A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares

m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")

plt.scatter(X, Y)
plt.plot(np.linspace(0, 5, 5), np.linspace(0, 5, 5)*m + c)
plt.show()
```

```python
"""
Part 2 Discussion:
While there is still a line of best fit, the randomized
nature of the vectors mean that sometimes the line of 
best fit not actually that close to fitting the data,
with little actual correlation.
"""
```

```python
# with random perturbation

M, B = 0.5, 3

X_b = np.arange(0, 1, 0.01)
Y_b = X_b*M + (np.random.uniform(0, 1, 100)+3)

A = np.vstack([X_b, np.ones(len(X_b))]).T
m2, c2 = np.linalg.lstsq(A, Y_b, rcond=None)[0]

X_actual, Y_actual = np.linspace(0, 1, 100), np.linspace(0, 1, 100)*m2 + c2

plt.scatter(X_b, Y_b)
plt.plot(X_actual, Y_actual)
plt.show()

residuals = np.abs(Y_actual - Y_b)
print(f"{residuals=}\n")
print(f"{max(residuals)=}, {min(residuals)=}")
```

```python
"""
Part 2 Discussion:
The Best fit line still fits the general upwards trend of the data,
with at at most +/- 0.52 difference between the best fit line and 
actual values.
"""
```
