```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 6, 40)
plt.subplot(2,2,1) # (row, col, active_area)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2,2,2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(2,2,3)
plt.plot(x+np.pi, np.cos(x+np.pi), 'b')
plt.subplot(2,2,4)
plt.plot(x*2, np.cos(x*2), 'y')
```
