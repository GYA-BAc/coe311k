```python
import numpy as np

def jacobi(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
import numpy as np

def gauss_seidel(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
import numpy as np

A = np.array([
    [10, 1, 2, 3, 1],
    [1, 12, 1, 2, 3],
    [2, 1, 14, 1, 2],
    [3, 2, 1, 15, 1],
    [1, 3, 2, 1, 16]
])

B = np.array([
    [2, 4, 5, 3, 6],
    [3, 1, 4, 5, 2],
    [6, 2, 3, 4, 1],
    [5, 3, 6, 2, 4],
    [4, 6, 2, 5, 3]
])
```

```python
# jacobi method

import time

x0 = np.random.randint(1000, size=(5, 5))

start = time.time()

solution, iterations = jacobi(A, B, x0)
end = time.time()

print(f"Jacobi Solution: {solution}")
print(f"Jacobi Iterations: {iterations}")
print(f"Execution took: {end - start} s")

```

```python
# gauss seidel method

x0 = np.random.randint(1000, size=(5, 5))

start = time.time()

solution, iterations = gauss_seidel(A, B, x0)
end = time.time()

print(f"GS Solution: {solution}")
print(f"GS Iterations: {iterations}")
print(f"Execution took: {end - start} s")
```
