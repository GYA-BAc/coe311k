```python
import numpy as np

def is_diag(matrix: np.array) -> bool:
    return all([matrix[i][i] > sum(row)-matrix[i][i] for i, row in enumerate(matrix)])
```

```python
A = np.array([
        [2,0,1],
        [1,3,1],
        [1,3,10]
])
print(f"Is the matrix A diagonally dominant? {is_diag(A)}")
print(A)
```

```python
B = np.array([
        [2,10,1],
        [1,3,1],
        [1,3,10]
])
print(f"Is the matrix A diagonally dominant? {is_diag(B)}")
print(B)
```
