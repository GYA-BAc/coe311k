```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
k = 1 # Constant in the equation (e.g., spring constant)
hs = [.01, .02, .03, .05, .1, .2, .3, .5] # Step size
x_start = 0 # Start of the interval
x_end = 10 # End of the interval
y_0 = 1.0 # Initial condition for y
v_0 = 0.0 # Initial condition for dy/dx


def run_iter(h):

    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    # Plot the results
    plt.plot(x_values, y_values, label=f'[h={h}] y(x)')
    plt.plot(x_values, v_values, label=f'[h={h}] v(x)',
    linestyle='--')
    
    
plt.figure(figsize=(10, 5))

for h in hs:
    run_iter(h)

plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.show()
```
