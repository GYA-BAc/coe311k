```python
X = np.sort(np.random.uniform(0, 1, 20))
Y = np.sort(np.random.uniform(0, 1, 20))

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

plt.scatter(X, Y)
plt.plot(np.linspace(0, 1, 5), np.linspace(0, 1, 5)*m + c)
plt.show()
```

```python
M, B = 0.5, 3

X_b = np.arange(0, 1, 0.01)
Y_b = X_b*M + (np.random.uniform(0, 1, 100)+3)

A = np.vstack([X_b, np.ones(len(X_b))]).T
m2, c2 = np.linalg.lstsq(A, Y_b, rcond=None)[0]

X_actual, Y_actual = np.linspace(0, 1, 100), np.linspace(0, 1, 100)*m2 + c2

plt.scatter(X_b, Y_b)
plt.plot(X_actual, Y_actual)
plt.show()

residuals = np.abs(Y_actual - Y_b)
print(f"{residuals=}\n")
print(f"{max(residuals)=}, {min(residuals)=}")
```
