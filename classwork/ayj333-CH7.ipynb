```python
import numpy as np

def jacobi(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
A_a = np.array([
 [0.582745, 0.48    , 0.10   , 0.      , 0.      ], 
 [0.48    , 1.044129, 0.46   , 0.10    , 0.      ],
 [0.10    , 0.46    , 1.10431, 0.44    , 0.10    ],
 [0.      , 0.10    , 0.44   , 0.963889, 0.42    ],
 [0.      , 0.      , 0.10   , 0.42    , 0.522565]], dtype=float)
b_a = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
```

```python
import time


x0 = np.zeros_like(b_a)

start = time.time()

solution, iterations = jacobi(A_a, b_a, x0)
end = time.time()

if (iterations != 100):    
    print(f"[Jacobi] Solution: {solution}")
    print(f"[Jacobi] Iterations: {iterations}")
    print(f"[Jacobi] Execution took: {end - start} s\n")
else:
    print("[Jacobi] No solution found within 100 iterations")
    
    
print()

x0 = np.zeros_like(b_a)

start = time.time()

solution, iterations = gauss_seidel(A_a, b_a, x0)
end = time.time()

if (iterations != 100):    
    print(f"[GS] Solution: {solution}")
    print(f"[GS] Iterations: {iterations}")
    print(f"[GS] Execution took: {end - start} s\n")
else:
    print("[GS] No solution found within 100 iterations")
```
