```python
import numpy as np
import matplotlib.pyplot as plt


def e_appx(step):
    # Define parameters
    h = step  # step size
    x_start, x_end = 0, 2
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    
    # Initial condition
    y_values[0] = 1  # y(0) = 1
    # Function representing dy/dx = -2y
    def f(x, y):
        return -2 * y

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
        
    y_err = np.abs(np.exp(x_values*-2) - y_values)
            
    # Plot the results
    plt.plot(x_values, y_values, label=f"step={step:.2f}")
    plt.plot(x_values, y_err, label=f"error")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()
    plt.show()
    
    
    return y_err
    
```

```python


for i in np.arange(0.01, 1, 0.05):
    err = e_appx(i)
    print(f"Total Error: {sum(err)}")
    print(f"================================")
    
```
