```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def sin_taylor(n, x):
    total = 0
    last_denom = 1
    for i in range(n):
        total += (-1)**i * x**(2*i+1) / last_denom
        last_denom *= (2*i+3)*(2*i+2)
    return total
```

```python
x = np.arange(0, 10, 0.1)

t_1 = np.array([sin_taylor(1, i) for i in x])
t_2 = np.array([sin_taylor(2, i) for i in x])
t_3 = np.array([sin_taylor(3, i) for i in x])
t_4 = np.array([sin_taylor(4, i) for i in x])

exact = np.sin(x)
```

```python
er_1 = abs(exact - t_1)
er_2 = abs(exact - t_2)
er_3 = abs(exact - t_3)
er_4 = abs(exact - t_4)
```

```python
plt.subplots_adjust(hspace=1)

plt.subplot(2,2,1) # (row, col, active_area)
plt.title("1 series")
plt.plot(x, er_1, 'r')
plt.subplot(2,2,2)
plt.title("2 series")
plt.plot(x, er_2, 'g')
plt.subplot(2,2,3)
plt.title("3 series")
plt.plot(x, er_3, 'b')
plt.subplot(2,2,4)
plt.title("4 series")
plt.plot(x, er_4, 'y')
```
