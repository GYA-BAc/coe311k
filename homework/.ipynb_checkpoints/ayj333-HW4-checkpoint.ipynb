```python
import numpy as np

def jacobi(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
import numpy as np

def gauss_seidel(A, b, x0, tol=1e-6, max_iterations=100):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
import numpy as np

A = np.array([
    [  3,-.1,-.2],
    [0.1,  7,-.3],
    [0.3,-.2, 10]
])

B = np.array([7.85, -19.3, 71.4])

x0 = np.array([0, 0, 0])
```

```python
# jacobi method


solution, iterations = jacobi(A, B, x0, tol=1e-9)

print(f"Jacobi Solution: {solution}")
print(f"Jacobi Iterations: {iterations}")

```

```python
# gs method


solution, iterations = gauss_seidel(A, B, x0, tol=1e-9)

print(f"GS Solution: {solution}")
print(f"GS Iterations: {iterations}")

```
