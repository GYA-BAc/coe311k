```python
import random
```

```python
# monte carlo

def a_pi(n):
    inside = 0
    for _ in range(0, n):
        inside += int((random.uniform(-1,1)**2 + random.uniform(-1,1)**2)**(1/2) < 1)
            
    return (4*inside/n)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#driver code

# create axes
x = np.arange(10, 10000, 20)
y = np.empty(500)

for (i, v) in enumerate(range(10, 10000, 20)):
    # fill error values
    y[i] = abs(a_pi(v) - np.pi)
```

```python
# plotting

plt.figure
plt.xlabel('samples')
plt.ylabel('error')
plt.plot(x, y, label="samples to error")
plt.legend()
plt.show()
```
